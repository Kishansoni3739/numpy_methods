```python
import numpy as np

students = np.array([
    [101, 78, 85, 90],
    [102, 60, -1, 75],
    [103, 88, 92, 95],
    [104, 55, 65, 120],
    [105, 80, 82, 84]
])
```

```python
student_ids = students[:, 0]
student_ids
```

```python
marks = students[:, 1:]
marks
```

```python
marks[(marks > 100) | ( marks < 0)] = 0
marks
```

```python
subjec_avg = np.mean(marks, axis = 0)
```

```python
subjec_avg
```

```python
valid_mark = np.where((marks == 0), subjec_avg, marks)
```

```python
valid_mark
```

```python
final = np.column_stack((student_ids, valid_mark))
final
```

```python
student_total = np.sum(final[:, 1:], axis = 1)
```

```python
student_total
```

```python
student_total_final = np.vstack((student_ids.T, student_total.T))
```

```python
idWise_Total = student_total_final.T
idWise_Total
```

```python
best_total_id = idWise_Total[np.argmax(idWise_Total[:, 1]),0]
print(best_total_id)
```

```python
student_ids = idWise_Total[:, 0]
total_marks = idWise_Total[:, 1]

import matplotlib.pyplot as plt

plt.figure()
plt.bar(student_ids, total_marks)
plt.xlabel("Student ID")
plt.ylabel("Total Marks")
plt.title("Student Total Marks")
plt.show()
```
